```python
import re
```

```python
cp = open('./cthulhu/characters/kanae/chatpalette.txt', 'r', encoding='utf-8').read()
```

```python
data = {}
skills = cp.split(sep = '\n')[1:-12]
for skill in skills:
    value      = re.sub(r"\D", "", skill.split(sep = ' ')[0])
    skill_name = re.sub("\【|\】", "", skill.split(sep = ' ')[1])
    data[skill_name] = int(value)
    
```
